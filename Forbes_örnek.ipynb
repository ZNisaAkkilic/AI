```python
df=df.rename(columns={"rank": "Sıra", "personName":"İsim","age":"Yaş","finalWorth": " Servet",
                      "category":"Kategori","country": "Ülke", "gender": "Cinsiyet"})
```

```python
## sıras ütnunu index yapma
df=df.set_index("Sıra")
```
